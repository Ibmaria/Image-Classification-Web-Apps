```python

from keras.preprocessing import image
from keras.models import Model, load_model
from keras.layers import Dense, GlobalAveragePooling2D, Flatten
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import RMSprop
import numpy as np
from keras.utils import plot_model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from keras.layers import Dropout
```

```python
from skimage import color
def preprocessing(image):
  
  image_to_gray = color.rgb2gray(image)
  gray_to_rgb = np.stack((image_to_gray,) * 3, axis=-1)
  #perform mean substraction
  image = np.array(gray_to_rgb) - np.array([103.939, 116.779, 123.68])
  return image

train_imageDatagen = ImageDataGenerator(preprocessing_function=preprocessing)

test_imageDatagen = ImageDataGenerator(preprocessing_function=preprocessing)

train_imageGenerator = train_imageDatagen.flow_from_directory('/gdrive/My Drive/CDIP Dataset/cvl/train/', batch_size=32, target_size=(229, 229))
test_imageGenerator = test_imageDatagen.flow_from_directory('/gdrive/My Drive/CDIP Dataset/cvl/test/', batch_size=32, target_size=(229, 229))
```

```python
base_model = InceptionV3(include_top=False, weights='imagenet', input_shape=(229, 229, 3))
#for layer in base_model.layers:
  #layer.trainable=False
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
#x = Dropout(0.2)(x)

predictions = Dense(4, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
plot_model(model, 'InceptionV3_.png', show_shapes=False, show_layer_names=True)
model.summary()
```

```python

model.compile(optimizer=RMSprop(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_imageGenerator, 
                               epochs=20,
                               validation_data=test_imageGenerator,
                               verbose=2)
```
